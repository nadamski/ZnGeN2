```python
def chem(mu):

    muZn = mu[0] + ZnTOTEN/2
    muGe = mu[1] + GeTOTEN/2
    muN  = mu[2] + N2TOTEN/2

    muO1 = ZnO_TOTEN/2 - muZn
    muO2 = GeO2_TOTEN/6 - muGe/2
    muO = min(muO1, muO2)

    global muS1, muS2, muS
    muS1  = ZnS_TOTEN / 4 -muZn
    muS2 = GeS2_TOTEN/32  -muGe/2
    muS = min(muS1,muS2)
    
    muSe = ZnSe_TOTEN / 4 - muZn
    
    muAl = AlN_TOTEN / 2 - muN
    muGa = GaN_TOTEN / 2 - muN
    muIn = InN_TOTEN / 2 - muN

    muH = H2TOTEN/2

    delta = 0.0025
    global Ef
    Ef = np.arange(0, CB-VB, delta)

    #Vacancies

    #V_N

    Nn3 = E_VNn3 - E0 + muN - 3 * (Ef + VB) + E_CN3
    Nn2 = E_VNn2 - E0 + muN - 2 * (Ef + VB) + E_CN2
    Nn1 = E_VNn1 - E0 + muN - 1 * (Ef + VB) + E_CN1
    N0 = E_VN0 - E0 + muN + 0 * (Ef + VB)
    N1 = E_VN1 - E0 + muN + 1 * (Ef + VB) + E_CN1
    N2 = E_VN2 - E0 + muN + 2 * (Ef + VB) + E_CN2
    N3 = E_VN3 - E0 + muN + 3 * (Ef + VB) + E_CN3
    global V_N 
    V_N = np.minimum.reduce([Nn3, Nn2, Nn1, N0, N1, N2, N3])
    global V_N_l
    V_N_l = [(E_VN1 + E_CN1 - E_VN3 - E_CN3)/2 - VB, 
            (E_VN0 - E_VN1 - E_CN1) - VB, 
            (E_VNn1 + E_CNn1 - E_VN0) - VB]

    #V_Zn

    Znp2 = E_VZnp2 - E0 + muZn + 2 * (Ef + VB) + E_CZnp2
    Znp1 = E_VZnp1 - E0 + muZn + 1 * (Ef + VB) + E_CZnp1
    Zn0 = E_VZn0 - E0 + muZn - 0 * (Ef + VB)
    Zn1 = E_VZn1 - E0 + muZn - 1 * (Ef + VB) + E_CZn1
    Zn2 = E_VZn2 - E0 + muZn - 2 * (Ef + VB) + E_CZn2
    global V_Zn
    V_Zn = np.minimum.reduce([Znp2, Znp1, Zn0,Zn1,Zn2])
    global V_Zn_l
    V_Zn_l = [(E_VZn0 - E_VZnp1 - E_CZnp1) - VB, 
            (E_VZn1 + E_CZn1 - E_VZn0 ) - VB, 
            (E_VZn2 + E_CZn2 - E_VZn1 - E_CZn1) - VB]

    #V_Ge

    Gep3 = E_VGep3 - E0 + muGe + 3 * (Ef + VB) + E_CGep3
    Gep2 = E_VGep2 - E0 + muGe + 2 * (Ef + VB) + E_CGep2
    Gep1 = E_VGep1 - E0 + muGe + 1 * (Ef + VB) + E_CGep1
    Ge0 = E_VGe0 - E0 + muGe - 0 * (Ef + VB) 
    Ge1 = E_VGe1 - E0 + muGe - 1 * (Ef + VB) + E_CGe1
    Ge2 = E_VGe2 - E0 + muGe - 2 * (Ef + VB) + E_CGe2
    Ge3 = E_VGe3 - E0 + muGe - 3 * (Ef + VB) + E_CGe3
    Ge4 = E_VGe4 - E0 + muGe - 4 * (Ef + VB) + E_CGe4
    global V_Ge 
    V_Ge = np.minimum.reduce([Gep3, Gep2, Gep1, Ge0, Ge1, Ge2, Ge3, Ge4])
    global V_Ge_l
    V_Ge_l = [(E_VGe1 + E_CGe1 - E_VGe0 ) - VB, 
            (E_VGe2 + E_CGe2 - E_VGe1 - E_CGe1) - VB,
            (E_VGe3 + E_CGe3 - E_VGe2 - E_CGe2) - VB]


    #Cation Antisites

    #Ge_Zn

    Ge_Znn3 = E_Ge_Znn3 - E0 - muGe + muZn - 3 * (Ef + VB) + E_CGe_Znn3
    Ge_Znn2 = E_Ge_Znn2 - E0 - muGe + muZn - 2 * (Ef + VB) + E_CGe_Znn2
    Ge_Znn1 = E_Ge_Znn1 - E0 - muGe + muZn - 1 * (Ef + VB) + E_CGe_Znn1
    Ge_Zn0 = E_Ge_Zn0 - E0 - muGe + muZn + 0 * (Ef + VB)
    Ge_Znp1 = E_Ge_Znp1 - E0 - muGe + muZn + 1 * (Ef + VB) + E_CGe_Znp1
    Ge_Znp2 = E_Ge_Znp2 - E0 - muGe + muZn + 2 * (Ef + VB) + E_CGe_Znp2
    Ge_Znp3 = E_Ge_Znp3 - E0 - muGe + muZn + 3 * (Ef + VB) + E_CGe_Znp3
    Ge_Znp4 = E_Ge_Znp4 - E0 - muGe + muZn + 4 * (Ef + VB) + E_CGe_Znp4
    global Ge_Zn 
    Ge_Zn = np.minimum.reduce([Ge_Znn3, Ge_Znn2, Ge_Znn1, Ge_Zn0, Ge_Znp1, Ge_Znp2, Ge_Znp3, Ge_Znp4])
    global Ge_Zn_l
    Ge_Zn_l = [(E_Ge_Zn0 - E_Ge_Znp2 - E_CGe_Znp2)/2 - VB]


    #Zn_Ge

    Zn_Gep2 = E_Zn_Gep2 - E0 + muGe - muZn + 2 * (Ef + VB) + E_CZn_Gep2
    Zn_Gep1 = E_Zn_Gep1 - E0 + muGe - muZn + 1 * (Ef + VB) + E_CZn_Gep1
    Zn_Ge0 = E_Zn_Ge0 - E0 + muGe - muZn - 0 * (Ef + VB)
    Zn_Ge1 = E_Zn_Ge1 - E0 + muGe - muZn - 1 * (Ef + VB) + E_CZn_Ge1
    Zn_Ge2 = E_Zn_Ge2 - E0 + muGe - muZn - 2 * (Ef + VB) + E_CZn_Ge2
    global Zn_Ge 
    Zn_Ge = np.minimum.reduce([Zn_Gep2, Zn_Gep1, Zn_Ge0, Zn_Ge1, Zn_Ge2])
    global Zn_Ge_l
    Zn_Ge_l = [(E_Zn_Ge0 - E_Zn_Gep1 - E_CZn_Gep1) - VB, 
            (E_Zn_Ge1 + E_CZn_Ge1 - E_Zn_Ge0 ) - VB, 
            (E_Zn_Ge2 + E_CZn_Ge2 - E_Zn_Ge1 - E_CZn_Ge1) - VB]
    
    ZnGePairn1 = E_ZnGePairn1 - E0 - 1 * (Ef + VB) + E_CZnGePairn1
    ZnGePair0 = E_ZnGePair0 - E0 + 0 * (Ef + VB)
    ZnGePairp1 = E_ZnGePairp1 - E0 + 1 * (Ef + VB) + E_CZnGePairp1
    
    global ZnGePair 
    ZnGePair = np.minimum.reduce([ZnGePairn1, ZnGePair0, ZnGePairp1])
    global ZnGePair_l
    ZnGePair_l = [(E_ZnGePair0 - E_ZnGePairp1 - E_CZnGePairp1) - VB, 
            (E_ZnGePairn1 + E_CZnGePairn1 - E_ZnGePair0 ) - VB]

    
    #Hydrogen Complexes

    V_ZnHn2 = E_VZnHn2 - E0 + muZn - muH - 2 * (Ef + VB) + E_CVZnHn2
    V_ZnHn1 = E_VZnHn1 - E0 + muZn - muH - 1 * (Ef + VB) + E_CVZnHn1
    V_ZnH0 = E_VZnH0 - E0 + muZn - muH - 0 * (Ef + VB)
    V_ZnHp1 = E_VZnHp1 - E0 + muZn - muH + 1 * (Ef + VB) + E_CVZnHp1
    
    global V_ZnH, V_ZnH_l
    V_ZnH = np.minimum.reduce([V_ZnHn2, V_ZnHn1, V_ZnH0, V_ZnHp1])
    V_ZnH_l = [(E_VZnHn2 + E_CVZnHn2 -E_VZnHn1 - E_CVZnHn1) - VB, 
            (E_VZnHn1 + E_CVZnHn1 - E_VZnH0) - VB, 
            (E_VZnH0 - E_VZnHp1 - E_CVZnHp1) - VB]

    V_NHn1 = E_VNHn1 - E0 + muN - muH - 1 * (Ef + VB) + E_CVNHn1
    V_NH0 = E_VNH0 - E0 + muN - muH - 0 * (Ef + VB)
    V_NHp1 = E_VNHp1 - E0 + muN - muH + 1 * (Ef + VB) + E_CVNHp1
    V_NHp2 = E_VNHp2 - E0 + muN - muH + 2 * (Ef + VB) + E_CVNHp2
    V_NHp3 = E_VNHp3 - E0 + muN - muH + 3 * (Ef + VB) + E_CVNHp3


    global V_NH, V_NH_l
    V_NH = np.minimum.reduce([V_NHn1, V_NH0, V_NHp1, V_NHp2, V_NHp3])
    V_NH_l = [(E_VNHn1 + E_CVNHn1 - E_VNH0) - VB, 
             (E_VNH0 - E_VNHp1 - E_CVNHp1) - VB, 
             (E_VNHp1 + E_CVNHp1 -E_VNHp2 - E_CVNHp2) - VB,
             (E_VNHp2 + E_CVNHp2 -E_VNHp3 - E_CVNHp3) - VB]

    


    #Nitrogen Antisites

    #Ge_N

    GeNn2 = E_Ge_Nn2 - E0 - muGe + muN - 2 * (Ef + VB) + E_CGe_N2
    GeNn1 = E_Ge_Nn1 - E0 - muGe + muN - 1 * (Ef + VB) + E_CGe_N1
    GeN0 = E_Ge_N0 - E0 - muGe + muN + 0 * (Ef + VB)
    GeN1 = E_Ge_N1 - E0 - muGe + muN + 1 * (Ef + VB) + E_CGe_N1
    GeN2 = E_Ge_N2 - E0 - muGe + muN + 2 * (Ef + VB) + E_CGe_N2

    #N_Ge

    NGen2 = E_N_Gen2 - E0 - muN + muGe - 2 * (Ef + VB) + E_CN_Gen2
    NGen1 = E_N_Gen1 - E0 - muN + muGe - 1 * (Ef + VB) + E_CN_Gen1
    NGe0 = E_N_Ge0 - E0 - muN + muGe + 0 * (Ef + VB)
    NGe1 = E_N_Ge1 - E0 - muN + muGe + 1 * (Ef + VB) + E_CN_Ge1
    NGe2 = E_N_Ge2 - E0 - muN + muGe + 2 * (Ef + VB) + E_CN_Ge2

    #Zn_N

    ZnNn1 = E_Zn_Nn1 - E0 - muZn + muN - 1 * (Ef + VB) + E_CZn_N1
    ZnN0 = E_Zn_N0 - E0 - muZn + muN + 0 * (Ef + VB)
    ZnN1 = E_Zn_N1 - E0 - muZn + muN + 1 * (Ef + VB) + E_CZn_N1

    #N_Zn

    NZnn1 = E_N_Znn1 - E0 + muZn - muN - 1 * (Ef + VB) + E_CN_Zn1
    NZn0 = E_N_Zn0 - E0 + muZn - muN + 0 * (Ef + VB)
    NZn1 = E_N_Zn1 - E0 + muZn - muN + 1 * (Ef + VB) + E_CN_Zn1

    #Interstitials

    #N_i

    Nip3 = E_Nip3 - E0 - muN + 3 * (Ef + VB) + E_CNip3
    Nip2 = E_Nip2 - E0 - muN + 2 * (Ef + VB) + E_CNip2
    Nip1 = E_Nip1 - E0 - muN + 1 * (Ef + VB) + E_CNip1
    Ni0 = E_Ni0 - E0 - muN + 0 * (Ef + VB) 
    Nin1 = E_Nin1 - E0 - muN - 1 * (Ef + VB) + E_CNin1
    Nin2 = E_Nin2 - E0 - muN - 2 * (Ef + VB) + E_CNin2
    global N_i 
    N_i = np.minimum.reduce([Nip3, Nip2, Nip1, Ni0, Nin1, Nin2])
    global N_i_l
    N_i_l = [(E_Nip2 + E_CNip2 - E_Nip3 - E_CNip3) - VB, 
            (E_Nip1 + E_CNip1 - E_Nip2 - E_CNip2) - VB, 
            (E_Ni0 - E_Nip1 - E_CNip1) - VB, 
            (E_Nin1 + E_CNin1 - E_Ni0) - VB]

    #Zn_i
    
    global Znip2, Znip1, Zni0
    Znip2 = E_Znip2 - E0 - muZn + 2 * (Ef + VB) + E_CZnip2 
    Znip1 = E_Znip1 - E0 - muZn + 1 * (Ef + VB) + E_CZnip1 
    Zni0 = E_Zni0 - E0 - muZn + 0 * (Ef + VB) 
    global Zn_i 
    Zn_i = np.minimum.reduce([Znip2, Znip1, Zni0]) 
    global Zn_i_l
    Zn_i_l = [(E_Znip1 + E_CZnip1 - E_Znip2 - E_CZnip2) - VB]
    
    #Ge_i
    
    global Geip2, Geip1, Gei0
    Geip4 = E_Geip4 - E0 - muGe + 4 * (Ef + VB) + E_CGeip4
    Geip3 = E_Geip3 - E0 - muGe + 3 * (Ef + VB) + E_CGeip3 
    Geip2 = E_Geip2 - E0 - muGe + 2 * (Ef + VB) + E_CGeip2
    Geip1 = E_Geip1 - E0 - muGe + 1 * (Ef + VB) + E_CGeip1 
    Gei0 = E_Gei0 - E0 - muGe + 0 * (Ef + VB) 
    global Ge_i 
    Ge_i = np.minimum.reduce([Geip4, Geip3, Geip2, Geip1, Gei0]) 
    global Ge_i_l
    Ge_i_l = [(E_Gei0 - E_Geip1 - E_CGeip1) - VB,
             (E_Geip1 + E_CGeip1 - E_Geip2 - E_CGeip2) - VB,
             (E_Geip2 + E_CGeip2 - E_Geip3 - E_CGeip3) - VB,
             (E_Geip3 + E_CGeip3 - E_Geip4 - E_CGeip4) - VB]
    

    #Unintentional Dopants

    #O_N
    
    global O_Nn1, O_N0, O_N1, Ef2
    #Ef2 = np.arange(0, 4, delta)
    O_Nn1 = E_O_Nn1 - E0 - muO + muN - 1 * (Ef + VB) + E_CO_Nn1
    O_N0 = E_O_N0 - E0 - muO + muN + 0 * (Ef + VB)
    O_N1 = E_O_N1 - E0 - muO + muN + 1 * (Ef + VB) + E_CO_N1
    O_N2 = E_O_N1 - E0 - muO + muN + 2 * (Ef + VB) + E_CO_N2
    global O_N 
    O_N = np.minimum.reduce([O_Nn1, O_N0, O_N1, O_N2])
    global O_N_l
    O_N_l = [(E_O_N0 - E_O_N1 - E_CO_N1) - VB, 
            (E_O_Nn1 + E_CO_Nn1 - E_O_N0) - VB]
    
    

    #H_i

    Hp1 = E_Hp1 - E0 - muH + 1 * (Ef + VB) + E_CHp1
    H0 = E_H0 - E0 - muH + 0 * (Ef + VB) 
    Hn1 = E_Hn1 - E0 - muH - 1 * (Ef + VB) + E_CHn1
    global H_i
    H_i = np.minimum.reduce([Hp1, H0, Hn1])
    global H_i_l
    H_i_l = [(E_Hn1 + E_CHn1 - E_Hp1 - E_CHp1)/2 - VB]
    
    
    #C_N


    C_Nn2 = E_C_Nn2 - E0 - muC + muN - 2 * (Ef + VB) + E_CC_Nn2
    C_Nn1 = E_C_Nn1 - E0 - muC + muN - 1 * (Ef + VB) + E_CC_Nn1
    C_N0 = E_C_N0 - E0 - muC + muN + 0 * (Ef + VB)
    C_Np1 = E_C_Np1 - E0 - muC + muN + 1 * (Ef + VB) + E_CC_Np1
    global C_N
    C_N = np.minimum.reduce([C_Nn2, C_Nn1, C_N0, C_Np1])


    #C_Ge
    
    global C_Gen1, C_Ge0, C_Gep1
    C_Gen1 = E_C_Gen1 - E0 - muC + muGe - 1 * (Ef + VB) + E_CC_Gen1
    C_Ge0 = E_C_Ge0 - E0 - muC + muGe + 0 * (Ef + VB)
    C_Gep1 = E_C_Gep1 - E0 - muC + muGe + 1 * (Ef + VB) + E_CC_Gep1
    global C_Ge
    C_Ge = np.minimum.reduce([C_Gen1, C_Ge0, C_Gep1])
    
    
    #Sn_Ge
    global Sn_Ge
    Sn_Ge0 = E_Sn_Ge0 - E0 - muSn + muGe + 0 * (Ef + VB)
    Sn_Ge = Sn_Ge0

    #Sn_Zn
    global Sn_Zn
    Sn_Zn0 = E_Sn_Zn0 - E0 - muSn + muZn + 0 * (Ef + VB)
    Sn_Znp2 = E_Sn_Znp2 - E0 - muSn + muZn + 2 * (Ef + VB) + E_CSn_Znp2
    Sn_Zn = np.minimum.reduce([Sn_Znp2, Sn_Zn0])
    
    
    #Intentional Dopants
    
    #Li 
    
    global Li_Znn1
    Li_Znn1 = E_Li_Znn1 - E0 - muLi + muZn - 1 * (Ef + VB) + E_CLi_Znn1
    global Li_Zn0
    Li_Zn0 = E_Li_Zn0 - E0 - muLi + muZn + 0 * (Ef + VB)
    global Li_Zn_l
    Li_Zn_l = [E_Li_Znn1 + E_CLi_Znn1 - E_Li_Zn0 - VB]
    global Li_Zn
    Li_Zn = np.minimum.reduce([Li_Znn1, Li_Zn0])


    #Cu
    
    global Cu_Znn1
    Cu_Znn1 = E_Cu_Znn1 - E0 - muCu + muZn - 1 * (Ef + VB) + E_CCu_Znn1
    global Cu_Zn0
    Cu_Zn0 = E_Cu_Zn0 - E0 - muCu + muZn + 0 * (Ef + VB)
    global Cu_Znp1
    Cu_Znp1 = E_Cu_Znp1 - E0 - muCu + muZn + 1 * (Ef + VB) + E_CCu_Znp1
    global Cu_Zn, Cu_Zn_l
    Cu_Zn_l = [E_Cu_Znn1 + E_CCu_Znn1 - E_Cu_Zn0 - VB,
              E_Cu_Zn0 - E_Cu_Znp1 - E_CCu_Znp1 - VB]
    Cu_Zn = np.minimum.reduce([Cu_Znn1, Cu_Zn0, Cu_Znp1])

    
    #Al
    
    Al_Zn0 = E_Al_Zn0 - E0 - muAl + muZn + 0 * (Ef + VB)
    Al_Znp1 = E_Al_Znp1 - E0 - muAl + muZn + 1 * (Ef + VB) + E_CAl_Znp1
    global Al_Zn, Al_Zn_l
    Al_Zn = np.minimum.reduce([Al_Zn0, Al_Znp1])
    Al_Zn_l = [E_Al_Zn0 - E_Al_Znp1 - E_CAl_Znp1 - VB]

    
    Al_Ge0 = E_Al_Ge0 - E0 - muAl + muGe - 0 * (Ef + VB)
    Al_Gen1 = E_Al_Gen1 - E0 - muAl + muGe - 1 * (Ef + VB) + E_CAl_Gen1
    global Al_Ge, Al_Ge_l
    Al_Ge = np.minimum.reduce([Al_Ge0, Al_Gen1])
    Al_Ge_l = [E_Al_Gen1 + E_CAl_Gen1 - E_Al_Ge0  - VB]
    
    AlH_Gep1 = E_AlH_Gep1 - E0 - muAl - muH + muGe + 1 * (Ef + VB) + E_CAlH_Gep1
    AlH_Ge0 = E_AlH_Ge0 - E0 - muAl - muH + muGe - 0 * (Ef + VB)
    AlH_Gen1 = E_AlH_Gen1 - E0 - muAl - muH + muGe - 1 * (Ef + VB) + E_CAlH_Gen1
    global AlH_Ge, AlH_Ge_l
    AlH_Ge = np.minimum.reduce([AlH_Ge0, AlH_Gen1])
    AlH_Ge_l = [E_AlH_Gen1 + E_CAlH_Gen1 - E_AlH_Ge0  - VB]
    #Ga
    
    Ga_Zn0 = E_Ga_Zn0 - E0 - muGa + muZn + 0 * (Ef + VB)
    Ga_Znp1 = E_Ga_Znp1 - E0 - muGa + muZn + 1 * (Ef + VB) + E_CGa_Znp1
    global Ga_Zn, Ga_Zn_l
    Ga_Zn = np.minimum.reduce([Ga_Zn0, Ga_Znp1])
    Ga_Zn_l = [E_Ga_Zn0 - E_Ga_Znp1 - E_CGa_Znp1 - VB]

    Ga_Ge0 = E_Ga_Ge0 - E0 - muGa + muGe - 0 * (Ef + VB)
    Ga_Gen1 = E_Ga_Gen1 - E0 - muGa + muGe - 1 * (Ef + VB) + E_CGa_Gen1
    global Ga_Ge, Ga_Ge_l
    Ga_Ge = np.minimum.reduce([Ga_Ge0, Ga_Gen1])
    Ga_Ge_l = [E_Ga_Gen1 + E_CGa_Gen1 - E_Ga_Ge0  - VB]

    
    GaH_Ge0 = E_GaH_Ge0 - E0 - muGa - muH + muGe - 0 * (Ef + VB)
    GaH_Gen1 = E_GaH_Gen1 - E0 - muGa - muH + muGe - 1 * (Ef + VB) + E_CGaH_Gen1
    global GaH_Ge, GaH_Ge_l
    GaH_Ge = np.minimum.reduce([GaH_Ge0, GaH_Gen1])
    GaH_Ge_l = [E_GaH_Gen1 + E_CGaH_Gen1 - E_GaH_Ge0  - VB]
    
    #In
    In_Zn0 = E_In_Zn0 - E0 - muIn + muZn + 0 * (Ef + VB)
    In_Znp1 = E_In_Znp1 - E0 - muIn + muZn + 1 * (Ef + VB) + E_CIn_Znp1
    global In_Zn, In_Zn_l
    In_Zn = np.minimum.reduce([In_Zn0, In_Znp1])
    In_Zn_l = [E_In_Zn0 - E_In_Znp1 - E_CIn_Znp1 - VB]

    In_Ge0 = E_In_Ge0 - E0 - muIn + muGe - 0 * (Ef + VB)
    In_Gen1 = E_In_Gen1 - E0 - muIn + muGe - 1 * (Ef + VB) + E_CIn_Gen1
    global In_Ge, In_Ge_l
    In_Ge = np.minimum.reduce([In_Ge0, In_Gen1])
    In_Ge_l = [E_In_Gen1 + E_CIn_Gen1 - E_In_Ge0  - VB]

    #S
    S_Nn1 = E_S_Nn1 - E0 - muS + muN - 1 * (Ef + VB) + E_CS_Nn1
    S_N0 = E_S_N0 - E0 - muS + muN + 0 * (Ef + VB)
    S_Np1 = E_S_N1 - E0 - muS + muN + 1 * (Ef + VB) + E_CS_N1
    global S_N, S_N_l
    S_N = np.minimum.reduce([S_Nn1, S_N0, S_Np1])
    S_N_l = [E_S_N0 - E_S_N1 - E_CS_N1 - VB]
    
    S_Gen1 = E_S_Gen1 - E0 - muS + muGe - 1 * (Ef + VB) + E_CS_Gen1
    S_Ge0 = E_S_Ge0 - E0 - muS + muGe + 0 * (Ef + VB)
    S_Gep1 = E_S_Ge1 - E0 - muS + muGe + 1 * (Ef + VB) + E_CS_Ge1
    global S_Ge, S_Ge_l
    S_Ge = np.minimum.reduce([S_Gen1, S_Ge0, S_Gep1])
    S_Ge_l = [E_S_Ge0 - E_S_Ge1 - E_CS_Ge1 - VB]
    
    S_Znn1 = E_S_Znn1 - E0 - muS + muGe - 1 * (Ef + VB) + E_CS_Znn1
    S_Zn0 = E_S_Zn0 - E0 - muS + muGe + 0 * (Ef + VB)
    S_Znp1 = E_S_Zn1 - E0 - muS + muGe + 1 * (Ef + VB) + E_CS_Zn1
    global S_Zn, S_Zn_l
    S_Zn = np.minimum.reduce([S_Znn1, S_Zn0, S_Znp1])
    S_Zn_l = [E_S_Zn0 - E_S_Zn1 - E_CS_Zn1 - VB]

    #Se
    Se_N0 = E_Se_N0 - E0 - muSe + muN + 0 * (Ef + VB)
    Se_Np1 = E_Se_N1 - E0 - muSe + muN + 1 * (Ef + VB) + E_CSe_N1
    global Se_N, Se_N_l
    Se_N = np.minimum.reduce([Se_N0, Se_Np1])
    Se_N_l = [E_Se_N0 - E_Se_N1 - E_CSe_N1 - VB]
    
    Se_Znn1 = E_Se_Znn1 - E0 - muSe + muZn - 1 * (Ef + VB) + E_CSe_Znn1
    Se_Zn0 = E_Se_Zn0 - E0 - muSe + muZn + 0 * (Ef + VB)
    Se_Znp1 = E_Se_Zn1 - E0 - muSe + muZn + 1 * (Ef + VB) + E_CSe_Zn1
    global Se_Zn, Se_Zn_l
    Se_Zn = np.minimum.reduce([Se_Znn1, Se_Zn0, Se_Znp1])
    Se_Zn_l = [E_Se_Zn0 - E_Se_Zn1 - E_CSe_Zn1 - VB]
    
    Se_Gen1 = E_Se_Gen1 - E0 - muSe + muZn - 1 * (Ef + VB) + E_CSe_Gen1
    Se_Ge0 = E_Se_Ge0 - E0 - muSe + muZn + 0 * (Ef + VB)
    Se_Gep1 = E_Se_Ge1 - E0 - muSe + muZn + 1 * (Ef + VB) + E_CSe_Ge1
    global Se_Ge, Se_Ge_l
    Se_Ge = np.minimum.reduce([Se_Gen1, Se_Ge0, Se_Gep1])
    Se_Ge_l = [E_Se_Ge0 - E_Se_Ge1 - E_CSe_Ge1 - VB]
```
